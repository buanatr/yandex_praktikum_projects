```python
def table_with_num_and_perc(df):
    data_skip = df.isnull().sum().sort_values()
    data_skip_perc = df.isnull().sum().sort_values()/df.shape[0]*100
    skip_table = pd.concat([data_skip, data_skip_perc], axis=1)
    return skip_table
```

Выбросы, настолько влияющие на статистику, безусловно нужно будет отсекать.

В столбцах, обозначающих количество комнат и высоту потолков таких выбросов не обнаружено. std = 1. Среднее значение по количеству комнат - 2 (впрочем, как и медианное). Средняя высота потолков, продаваемых квартир равна 2,7 метра.

Для того, чтобы доработать предыдущую инструкцию (описать, сколько обычно занимает продажа, когда можно считать, что продажи прошли очень быстро, а когда необычно долго), а также для отработки следующей инструкции (убрать редкие и выбивающиеся значения, описать, какие особенности обнаружили) - напишем функцию, отсекающую выбросы. Также с помощью следующей функции мы сможем получить адекватные значения средних, медианных и прочих величин.

Основная структура функции подсмотрена на сайте "https://stackoverrun.com/ru/q/6353870"

```python
def remove_outlier(df, col_name):
       q1 = df[col_name].quantile(0.25)
       q3 = df[col_name].quantile(0.75)
       iqr = q3-q1 
       fence_low  = q1-1.5*iqr
       fence_high = q3+1.5*iqr
       df_out = df.loc[(df[col_name] > fence_low) & (df[col_name] < fence_high)]
       return df_out[col_name]
```

```python
histos = remove_outlier(df, 'days_exposition').hist(bins=100, range=(1, 520))
plt.title('Распределение времени продажи квартир (без выбросов)')
histos.set_xlabel('Время продажи (в днях)')
histos.set_ylabel('Частота встречаемости (в количестве объявлений)')
```

```python
def remove_outlier(df, col_name):
       q1 = df[col_name].quantile(0.25)
       q3 = df[col_name].quantile(0.75)
       iqr = q3-q1 
       fence_low  = q1-1.5*iqr
       fence_high = q3+1.5*iqr
       df_out = df.loc[(df[col_name] > fence_low) & (df[col_name] < fence_high)]
       return df_out[col_name]
```
